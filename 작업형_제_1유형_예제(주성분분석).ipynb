```r
#mtcars 데이터셋에서 mpg변수를 제외하고 데이터 정규화 (standardscaler) 과정을 진행한 이후 
#PCA를 통해 변수 축소를 하려한다. 누적설명 분산량이 92%를 넘기기 위해서는 몇개의 주성분을 
#선택해야하는지 설명하라

df<- read.csv("https://raw.githubusercontent.com/Datamanim/dataq/main/mtcars.csv",row.names=1)
#head(df)
df_scale <- scale(df)
pca <- prcomp(df_scale)
summary(pca)
#PC4
```

```r
#mtcars 의 index는 (업체명) - (모델명)으로 구성된다. (valiant는 업체명) mtcars에 ‘brand’ 
#컬럼을 추가하고 value 값으로 업체명을 입력하라

#추가된 brand 컬럼을 제외한 모든 컬럼을 통해 pca를 실시한다. 2개의 주성분과 brand컬럼으로 구성된 
#새로운 데이터 프레임을 출력하고, brand에 따른 2개 주성분을 시각화하여라 (brand를 구분 할수 있도록 
#색이다른 scatterplot, legend를 표시한다)
df<- read.csv("https://raw.githubusercontent.com/Datamanim/dataq/main/mtcars.csv",row.names=1)
b<-sapply(strsplit(rownames(df)," "),`[`,1)
df$brand<-b
df_scale<-scale(df[,-12])
pca<-prcomp(df_scale)
df <- cbind(df, pca$x[,1:2])
#install.packages("ggplot2")
library(ggplot2)
ggplot(df,aes(x=PC1,y=PC2,color=brand))+geom_point()

```
