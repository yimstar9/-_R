```r
### 주어진 데이터(basic2.csv)에서 주 단위 Sales의 합계를 구하고, 가장 큰 값을 가진 주와 작은 값을 가진 주의 차이를 구하시오(절대값)
### 데이터셋 : basic2.csv
library(dplyr)
library(lubridate)
df<-read.csv('basic2.csv')
df%>%mutate(w=week(Date))%>%group_by(w)%>%summarise(v=sum(Sales))%>%select(v)%>%summarise(ans=max(v)-min(v))

library(readr)
df%>%mutate(week=cut(Date,breaks='week'))%>% group_by(week) %>% summarise(value=sum(Sales)) %>% summarise(ans=max(value)-min(value))
```
